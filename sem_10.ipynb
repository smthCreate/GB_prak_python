```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
df.describe()
```

```python
sns.scatterplot(data=df,x='households',y='population')
```

```python
sns.relplot(x='longitude',y='median_house_value',kind ='line',data=df)
```

```python
sns.histplot(data=df,x="housing_median_age")
```

```python
sns.histplot(data=df[(df['housing_median_age']>40)&(df['housing_median_age']<51)],x="median_income")
```

```python
sns.relplot(x='housing_median_age',y='median_income',kind ='line',data=df)
```
